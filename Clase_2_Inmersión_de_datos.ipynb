Devolverá una separación

Con expand = True nós devolverá un dataframe con dos columnas

```python
valor = inmuebles.Valor.str.split(expand= True)

inmuebles["Moneda"] = valor[0]
inmuebles["Precio"] = valor[1]

inmuebles.sample(3)
```

```python
inmuebles["Precio"] = inmuebles["Precio"].str.replace(".","",regex=True)

inmuebles[["Precio","Barrio"]]
```

```python
inmuebles.info()
```

Transformar a tipo númerico la columna Precio

```python
inmuebles["Precio_Millon"] = inmuebles.Precio.astype("float")/1000000
inmuebles.info()
```

Graficando el precio/M2

---



```python
inmuebles["Precio_M2"] = inmuebles["Precio_Millon"]/inmuebles["Área"]
inmuebles.info()
```

```python
inmuebles.groupby("Barrio").mean()
```

```python
datos_Barrio = inmuebles.groupby("Barrio").sum()
datos_Barrio
```

```python
datos_Barrio["Precio_M2_Barrio"] = datos_Barrio["Precio_Millon"]/datos_Barrio["Área"]
datos_Barrio
```

```python
m2_Barrio = dict(datos_Barrio["Precio_M2_Barrio"])
m2_Barrio
```

```python
inmuebles["Precio_M2_Barrio"] = inmuebles["Barrio"]
inmuebles["Precio_M2_Barrio"] = inmuebles["Precio_M2_Barrio"].map(m2_Barrio)
inmuebles
```

```python
top_Barrios = inmuebles["Barrio"].value_counts()[:10].index
top_Barrios
```

```python
datos_Barrio.reset_index(inplace = True)
datos_Barrio
```

```python
datos_Barrio.query("Barrio in @top_Barrios")
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x="Barrio", y = "Precio_M2_Barrio", data = datos_Barrio.query("Barrio in @top_Barrios"))
ax.tick_params(axis = "x", rotation = 45)
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y = "Precio_M2", data = inmuebles.query("Barrio in @top_Barrios"))
ax.tick_params(axis = "x", rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y = "Precio_M2", data = inmuebles.query("Barrio in @top_Barrios & Precio_M2 < 15"))
ax.tick_params(axis = "x", rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y = "Área", data = inmuebles.query("Barrio in @top_Barrios & Área < 500"))
ax.tick_params(axis = "x", rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y = "Precio_Millon", data = inmuebles.query("Barrio in @top_Barrios & Precio_Millon < 2000"))
ax.tick_params(axis = "x", rotation = 45)
plt.show()
```
